# e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month?
(Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment
– Monthly spend. Interest is earned only on positive profits and not on negative amounts

```python
#merging all the three tables.
```

```python
cus_acqusition_spend_repay = pd.merge(left=cus_acqusition_spend,right=repay,on="Customer",how="inner")
```

```python
cus_acqusition_spend_repay
```

```python
# renaming the columns for clearity
```

```python
cus_acqusition_spend_repay.rename(columns={"Amount_x":"Spend_Amount","Amount_y":"Repay_Amount"},inplace=True)
```

```python
cus_acqusition_spend_repay.head()
```

```python
# grouping the data based on "Yearly","Month_x" columns to get the 'Spend_Amount'and 'Repay_Amount
```

```python
interest_group = cus_acqusition_spend_repay.groupby(["Yearly","Monthly"])['Spend_Amount','Repay_Amount'].sum()
```

```python
interest_group
```

```python
 # Monthly Profit = Monthly repayment – Monthly spend
```

```python
interest_group['Monthly Profit'] = interest_group['Repay_Amount'] - interest_group['Spend_Amount']
```

```python
interest_group
```

```python
#interest earned is 2.9% of Monthly Profit
```

```python
interest_group['Interest Earned'] = (2.9* interest_group['Monthly Profit'])/100
```

```python
interest_group
```

# 5. Write user defined PYTHON function to perform the following analysis:
You need to find top 10 customers for each city in terms of their repayment amount by
different products and by different time periods i.e. year or month. The user should be able
to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the
function should automatically take these inputs while identifying the top 10 customers

```python
cus_repay.head()
```

```python
# converting 'Month' column to date time format
```

```python
cus_repay['Month'] = pd.to_datetime(cus_repay['Month'])
```

```python
#creating new column "Monthly" and "Yearly" using already existing 'Month' column
```

```python
cus_repay['Monthly'] = cus_repay['Month'].apply(lambda x:pd.Timestamp.strftime(x,format="%B"))
```

```python
cus_repay['Yearly'] = cus_repay['Month'].apply(lambda x:pd.Timestamp.strftime(x,format="%Y"))
```

```python
def top10Customers(product_category,time_period):
    return cus_repay.loc[(cus_repay.Product == product_category)].groupby(['Customer','City','Product',time_period]).Amount.sum().reset_index().sort_values('Amount',ascending=False).head(10)
```

```python
product_category=str(input("Please Enter Product Category and product category should be in Gold/Silver/Platinum: "))
```

```python
time_period=str(input("Please Enter Time Period and time period should be in yearly/monthly: "))
```

```python
top10Customers_Analysis=top10Customers(product_category,time_period)
```
